```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine, Table, Column, MetaData
from sqlalchemy import Integer, String, Float, Boolean

# Download the dataset from Hugging Face
print("Downloading dataset...")
df = pd.read_csv("hf://datasets/maharshipandya/spotify-tracks-dataset/dataset.csv")
print(f"Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"Columns: {list(df.columns)}")
print(f"\nData types:\n{df.dtypes}")

# PostgreSQL connection parameters (replace these with your actual values)
DB_HOST = "194.147.142.26"
DB_PORT = "5432"
DB_NAME = "dsdb"
DB_USER = "niflheim"
DB_PASSWORD = "niflguard"
TABLE_NAME = "spotify_tracks"

# Create connection string
connection_string = f"postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"

# Create SQLAlchemy engine
print("\nConnecting to PostgreSQL...")
engine = create_engine(connection_string)

# Create table (pandas will do this automatically based on DataFrame structure)
print(f"Creating/replacing table '{TABLE_NAME}'...")
df.to_sql(
    name=TABLE_NAME,
    con=engine,
    if_exists='replace',
    index=False,
    chunksize=1000,
    method='multi'
)

print(f"Successfully created table and loaded {df.shape[0]} rows into '{TABLE_NAME}'")

# Verify the data was loaded
with engine.connect() as conn:
    result = conn.execute(sqlalchemy.text(f"SELECT COUNT(*) FROM {TABLE_NAME}"))
    count = result.fetchone()[0]
    print(f"Verification: Table contains {count} rows")

engine.dispose()
print("Done!")
```
